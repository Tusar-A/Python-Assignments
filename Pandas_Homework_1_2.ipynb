```python
import pandas as pd
```

```python
# Reading dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head(2) # viewing top two rows of the dataset
```

```python
# find out how many male and female passengers was onboarded.

male_members = len(df[df['Sex']== 'male'])
female_members =len(df[df['Sex']== 'female'])
print("Male passengers: {} \nFemale passengers: {}".format(male_members, female_members))
```

```python
# how many survived we have
survived_ = df['Survived'].value_counts()[0]
print("Survived: ", survived_)
```

```python
# what is name of a person who is the eldest one

eldest_person = df.loc[df['Age'] == df['Age'].max()]['Name']

print("Name of the eldest: ", eldest_person.values[0])
```

```python
# how many passenger do we have in first, second and third class
def p_class(data, col_name, p_class):
    num_passenger = len(data[data[col_name] == p_class])
    return "Number of passenger in class {} is {}".format(p_class, num_passenger)
    
print(p_class(data=df, col_name='Pclass', p_class= 1))
print(p_class(data=df, col_name='Pclass', p_class= 2))
print(p_class(data=df, col_name='Pclass', p_class= 3))
```

```python
# how many person we have whosee name starts with "s"

name_starts_with_s = len(df[df['Name'] == df['Name'].str.startswith("S")])
print(f"{name_starts_with_s} people's name starts with s")
```

```python
# try to create a column which is a summation of "sibsp" and "parch"

df['Sum_Sib_par'] = df['SibSp'] + df['Parch']
df.head(2)
```

```python
# how many person do we have below age of 25

age_under_25 = len(df[df['Age'] <25])
print("Number of people under 25: ", age_under_25)
```

```python
# how many person died whose age was less than 40

died_under_40 = len(df[(df['Age'] <40) & (df['Survived'] == 0)])
print("Number of person died under 40 are: ", died_under_40)
```
